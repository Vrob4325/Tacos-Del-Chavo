```python
# ========================================
# AN√ÅLISIS TACOS DEL CHAVO - VERSI√ìN COMPLETA
# ========================================

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

print("=" * 60)
print("üåÆ AUT√âNTICOS TACOS AL VAPOR EL CHAVO")
print("üìä Sistema de An√°lisis de Datos")
print("=" * 60)

# ========================================
# 1. CARGAR TODOS LOS ARCHIVOS
# ========================================
print("\nüìÅ Cargando datos...")

clientes = pd.read_csv('Clientes.csv')
productos = pd.read_csv('Productos.csv')
pedidos = pd.read_csv('Pedidos.csv')
inventario = pd.read_csv('Inventario.csv')
eventos = pd.read_csv('Eventos.csv')

# Limpiar nombres de columnas (quitar espacios)
clientes.columns = clientes.columns.str.strip()
productos.columns = productos.columns.str.strip()
pedidos.columns = pedidos.columns.str.strip()
inventario.columns = inventario.columns.str.strip()
eventos.columns = eventos.columns.str.strip()

print("‚úÖ Todos los datos cargados correctamente!")

# ========================================
# 2. MOSTRAR DATOS
# ========================================
print("\n" + "=" * 60)
print("üìã DATOS DEL SISTEMA")
print("=" * 60)

print("\nüë• CLIENTES:")
print(clientes)
print(f"total: {len(clientes)} clientes")

print("\nüåÆ PRODUCTOS:")
print(productos)
print(f"total: {len(productos)} productos")

print("\nüì¶ PEDIDOS:")
print(pedidos)
print(f"total: {len(pedidos)} pedidos")

print("\nüìä INVENTARIO:")
print(inventario)

print("\nüéâ EVENTOS:")
print(eventos)

# ========================================
# 3. ESTAD√çSTICAS
# ========================================
print("\n" + "=" * 60)
print("üìà ESTAD√çSTICAS")
print("=" * 60)

ventas_totales = pedidos['total'].sum()
ticket_promedio = pedidos['total'].mean()

print(f"\nüí∞ Ventas Totales: ${ventas_totales:,.2f} MXN")
print(f"üßæ Ticket Promedio: ${ticket_promedio:,.2f} MXN")
print(f"\nüìä Pedidos por tipo:")
print(pedidos['tipoPedido'].value_counts())

# ========================================
# 4. GR√ÅFICAS PRINCIPALES
# ========================================
print("\n" + "=" * 60)
print("üìä GENERANDO GR√ÅFICAS...")
print("=" * 60)

# Crear figura con 6 gr√°ficas
fig = plt.figure(figsize=(18, 12))

# GR√ÅFICA 1: Distribuci√≥n de Clientes
plt.subplot(2, 3, 1)
clientes['TipoCliente'].value_counts().plot(kind='bar',
    color=['#FF6B6B', '#4ECDC4', '#45B7D1'])
plt.title('Distribuci√≥n de Clientes por Tipo', fontsize=14, fontweight='bold')
plt.xlabel('Tipo de Cliente')
plt.ylabel('Cantidad')
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.3)

# GR√ÅFICA 2: Pedidos por Tipo (Pie)
plt.subplot(2, 3, 2)
pedidos['tipoPedido'].value_counts().plot(kind='pie', autopct='%1.1f%%',
    colors=['#FFB6B9', '#FEC8D8', '#957DAD'], startangle=90)
plt.title('Pedidos por Tipo de Servicio', fontsize=14, fontweight='bold')
plt.ylabel('')

# GR√ÅFICA 3: M√©todos de Pago
plt.subplot(2, 3, 3)
pedidos['metodoPago'].value_counts().plot(kind='barh',
    color=['#95E1D3', '#F38181', '#EAFFD0'])
plt.title('M√©todos de Pago', fontsize=14, fontweight='bold')
plt.xlabel('Cantidad')
plt.grid(axis='x', alpha=0.3)

# GR√ÅFICA 4: Ventas por Tipo de Pedido
plt.subplot(2, 3, 4)
ventas_tipo = pedidos.groupby('tipoPedido')['total'].sum()
ventas_tipo.plot(kind='bar', color=['#A8E6CF', '#FFD3B6', '#FFAAA5'])
plt.title('Ventas Totales por Tipo', fontsize=14, fontweight='bold')
plt.xlabel('Tipo de Pedido')
plt.ylabel('Ventas (MXN)')
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.3)

# GR√ÅFICA 5: Productos por Categor√≠a
plt.subplot(2, 3, 5)
productos['categoria'].value_counts().plot(kind='pie', autopct='%1.1f%%',
    colors=['#FFB6B9', '#FEC8D8', '#957DAD', '#D291BC'], startangle=90)
plt.title('Productos por Categor√≠a', fontsize=14, fontweight='bold')
plt.ylabel('')

# GR√ÅFICA 6: Estado del Inventario
plt.subplot(2, 3, 6)
inv_critico = inventario[inventario['stockActual'] < inventario['stockMinimo']]
inv_normal = inventario[inventario['stockActual'] >= inventario['stockMinimo']]

categorias = ['Stock Normal', 'Stock Cr√≠tico']
valores = [len(inv_normal), len(inv_critico)]
colores = ['#90EE90', '#FF6B6B']

plt.bar(categorias, valores, color=colores)
plt.title('Estado del Inventario', fontsize=14, fontweight='bold')
plt.ylabel('Cantidad de Productos')
plt.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.savefig('graficas_tacos_chavo.png', dpi=300, bbox_inches='tight')
print("\n‚úÖ Gr√°ficas generadas exitosamente!")
plt.show()

# ========================================
# 5. GR√ÅFICAS ADICIONALES
# ========================================
print("\nüìä Generando gr√°ficas adicionales...")

fig2, axes = plt.subplots(2, 2, figsize=(16, 10))

# GR√ÅFICA 7: Estado de Pedidos
pedidos['estadoPedido'].value_counts().plot(kind='barh', ax=axes[0, 0],
    color=['#90EE90', '#FFD93D', '#6BCB77', '#FF6B6B'])
axes[0, 0].set_title('Estado Actual de Pedidos', fontsize=14, fontweight='bold')
axes[0, 0].set_xlabel('Cantidad')
axes[0, 0].grid(axis='x', alpha=0.3)

# GR√ÅFICA 8: Top Productos M√°s Caros
top_productos = productos.nlargest(5, 'precio')[['nombre', 'precio']]
axes[0, 1].barh(top_productos['nombre'], top_productos['precio'], color='#95E1D3')
axes[0, 1].set_title('Top Productos M√°s Caros', fontsize=14, fontweight='bold')
axes[0, 1].set_xlabel('precio (MXN)')
axes[0, 1].grid(axis='x', alpha=0.3)

# GR√ÅFICA 9: Tipos de Eventos
eventos['tipoEvento'].value_counts().plot(kind='bar', ax=axes[1, 0],
    color=['#FFB6B9', '#FEC8D8', '#957DAD'])
axes[1, 0].set_title('Tipos de Eventos', fontsize=14, fontweight='bold')
axes[1, 0].set_xlabel('Tipo')
axes[1, 0].set_ylabel('Cantidad')
axes[1, 0].tick_params(axis='x', rotation=45)
axes[1, 0].grid(axis='y', alpha=0.3)

# GR√ÅFICA 10: Stock Actual vs M√≠nimo
axes[1, 1].bar(range(len(inventario)), inventario['stockActual'],
               label='Stock Actual', color='#4ECDC4', alpha=0.7)
axes[1, 1].bar(range(len(inventario)), inventario['stockMinimo'],
               label='Stock M√≠nimo', color='#FF6B6B', alpha=0.7)
axes[1, 1].set_title('Inventario: Actual vs M√≠nimo', fontsize=14, fontweight='bold')
axes[1, 1].set_xlabel('Producto ID')
axes[1, 1].set_ylabel('Cantidad')
axes[1, 1].legend()
axes[1, 1].grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.savefig('graficas_adicionales_tacos.png', dpi=300, bbox_inches='tight')
print("‚úÖ Gr√°ficas adicionales generadas!")
plt.show()

# ========================================
# 6. REPORTE FINAL
# ========================================
print("\n" + "=" * 60)
print("üìã REPORTE FINAL")
print("=" * 60)

print(f"""
üéØ RESUMEN EJECUTIVO:
   ‚Ä¢ total de Clientes: {len(clientes)}
   ‚Ä¢ total de Productos: {len(productos)}
   ‚Ä¢ total de Pedidos: {len(pedidos)}
   ‚Ä¢ total de Eventos: {len(eventos)}

üí∞ VENTAS:
   ‚Ä¢ Ventas Totales: ${ventas_totales:,.2f} MXN
   ‚Ä¢ Ticket Promedio: ${ticket_promedio:,.2f} MXN
   ‚Ä¢ Pedido M√°ximo: ${pedidos['total'].max():,.2f} MXN
   ‚Ä¢ Pedido M√≠nimo: ${pedidos['total'].min():,.2f} MXN

üì¶ INVENTARIO:
   ‚Ä¢ Productos con Stock Normal: {len(inv_normal)}
   ‚Ä¢ Productos con Stock Cr√≠tico: {len(inv_critico)}
   {'   ‚ö†Ô∏è ALERTA: Reabastecer productos' if len(inv_critico) > 0 else '   ‚úÖ Stock adecuado'}

üéâ EVENTOS:
   ‚Ä¢ Completados: {len(eventos[eventos['estadoEvento'] == 'Completado'])}
   ‚Ä¢ Confirmados: {len(eventos[eventos['estadoEvento'] == 'Confirmado'])}
""")

print("=" * 60)
print("‚úÖ AN√ÅLISIS COMPLETADO")
print("=" * 60)
print("\nüéì Proyecto: Aut√©nticos Tacos al Vapor El Chavo")
print("üìç Los Mochis, Sinaloa, M√©xico")
print("üìä Sistema de Base de Datos y An√°lisis")
```
